# 04 - Composition Analysis

**Goal:** does nanoplastic exposure change *how many* of each cell type there are? A drop in one population or a rise in another is the first hint of an immune effect.

We compute the proportion of each cell type within each of the four samples and compare them side by side.

```python
import sys, os
sys.path.append(os.path.abspath(".."))   # make the src/ package importable
from src import config as cfg
from src import utils as ut
ut.setup_scanpy()

import scanpy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("scanpy", sc.__version__)
```

```python
adata = ut.load_checkpoint('03_annotated')
```

## Counts and proportions table

```python
counts = pd.crosstab(adata.obs['sample'], adata.obs['cell_type'])
props = counts.div(counts.sum(axis=1), axis=0)
props.to_csv(cfg.TABLE_DIR / '04_composition_proportions.csv')
props
```

## Stacked bar — composition per sample

```python
ax = props.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='tab20')
ax.set_ylabel('proportion of cells'); ax.set_xlabel('sample')
ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=8)
plt.tight_layout(); plt.savefig(cfg.FIG_DIR / '04_composition_stacked.png', dpi=150); plt.show()
```

## Per-cell-type bars relative to control
Easier to read which populations move with exposure.

```python
rel = props.div(props.loc[cfg.CONTROL_LABEL])
ax = rel.T.plot(kind='bar', figsize=(12, 5))
ax.axhline(1.0, color='k', ls='--', lw=1)
ax.set_ylabel('proportion relative to control')
plt.tight_layout(); plt.savefig(cfg.FIG_DIR / '04_composition_relative.png', dpi=150); plt.show()
```
